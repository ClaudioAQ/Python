**Adicionando Marcadores**

```
mapa = folium.Map(location=[-8.0538900, -34.8811100],
           tiles="Stamen Terrain",
           zoom_start=14
           )
```

```
folium.Marker([-8.0631,  -34.8713],
              popup='<i>Marco zero</i>',
              tooltip="Clique Aqui").add_to(mapa)
```

```
mapa
```

```
folium.Marker([-8.0631,  -34.8713],
              popup='<i>Marco zero</i>',
              tooltip="Clique Aqui",
              icon=folium.Icon(color='green')).add_to(mapa)
```

```
mapa
```

```
folium.CircleMarker(location=[-8.0631,  -34.8713],
                    radius=50,
                    color='#3186cc',
                    fill=True,
                    fill_color='#3186cc').add_to(mapa)
```

```
mapa
```

```
import pandas as pd
from folium import plugins
```

```
from google.colab import files
arquivo = files.upload()
```

```
df = pd.read_csv("olist_geolocation_dataset.csv")
```

```
df.head()
```

```
df.info()
```

```
coordenadas = []
for lat,lng in zip(df.geolocation_lat.values[:18000], df.geolocation_lng.values[:18000]):
  coordenadas.append([lat,lng])
```

```
mapa = folium.Map(location=[-8.0538900, -34.8811100], zoom_start=5, tiles="Stamen Toner")
```

```
mapa.add_child(plugins.HeatMap(coordenadas))
mapa
```

Visualizando os dados mais distribuídos no mapa

```
df.geolocation_state.unique()
```

```
df.geolocation_state.value_counts()
```

```
df2 = df.sample(frac=0.03)
```

```
coordenadas = []
for lat,lng in zip(df2.geolocation_lat.values[:18000], df2.geolocation_lng.values[:18000]):
  coordenadas.append([lat,lng])
```

```
mapa = folium.Map(location=[-8.0538900, -34.8811100], zoom_start=8, tiles="Stamen Terrain")
```

```
mapa.add_child(plugins.HeatMap(coordenadas))
mapa
```
