# grouped calculations

```python
df.head(10)
```

```python
#Expectativa de vida para cada ano
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
df.groupby("continent")["country"].count()
```

```python
oceania = df.loc[df["continent"] == "Oceania"]
oceania
```

```python
df\
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]\
    .mean()
```

```python
(df
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]
    .mean()
    .reset_index()
)
```

```python
df.groupby("continent")["country"].nunique()
```

```python
df.groupby("continent")["country"].value_counts()
```

```python
pais = df.loc[df["country"] == "Angola"]
pais
```

```python
#Criando Gráficos com a biblioteca Matplotlib
import matplotlib.pyplot as plt
# probably don't need the magic anymore
%matplotlib inline

(df
     .groupby(['year'])
     [['lifeExp']]
     .mean()
     .plot()
)
plt.show()
```
